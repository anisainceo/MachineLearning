```python
# Import necessary libraries
import ydf
import numpy as np
import pandas as pd

# Load the dataset
path = "https://storage.googleapis.com/download.tensorflow.org/data/palmer_penguins/penguins.csv"
dataset = pd.read_csv(path)
label = "species"

# Display the first 3 examples
print(dataset.head(3))

# Split the dataset into training and testing sets
np.random.seed(1)
is_test = np.random.rand(len(dataset)) < 0.2

train_dataset = dataset[~is_test]
test_dataset = dataset[is_test]

print("Training examples: ", len(train_dataset))
print("Testing examples: ", len(test_dataset))

# Train the initial model and evaluate its performance
model = ydf.CartLearner(label=label).train(train_dataset)
model.plot_tree()
train_evaluation = model.evaluate(train_dataset)
print("Train accuracy:", train_evaluation.accuracy)

test_evaluation = model.evaluate(test_dataset)
print("Test accuracy:", test_evaluation.accuracy)

# Optimize the model by adjusting min_examples and max_depth parameters

# Trying different values for min_examples
print("Optimizing min_examples parameter:")
for min_examples in [1, 5, 10]:
    model_min_examples = ydf.CartLearner(label=label, min_examples=min_examples).train(train_dataset)
    model_min_examples.plot_tree()
    min_examples_evaluation = model_min_examples.evaluate(test_dataset)
    print(f"Test accuracy with min_examples={min_examples}: {min_examples_evaluation.accuracy}")

# Trying different values for max_depth
print("\nOptimizing max_depth parameter:")
for max_depth in [3, 5, 10]:
    model_max_depth = ydf.CartLearner(label=label, max_depth=max_depth).train(train_dataset)
    model_max_depth.plot_tree()
    max_depth_evaluation = model_max_depth.evaluate(test_dataset)
    print(f"Test accuracy with max_depth={max_depth}: {max_depth_evaluation.accuracy}")
```

```python
# Import necessary libraries
import numpy as np
import pandas as pd
import tensorflow_decision_forests as tfdf
from sklearn.model_selection import train_test_split

# Load the dataset
path = "https://storage.googleapis.com/download.tensorflow.org/data/palmer_penguins/penguins.csv"
dataset = pd.read_csv(path)
label = "species"

# Display the first 3 examples
print(dataset.head(3))

# Preprocess the data: drop rows with missing values
dataset = dataset.dropna()

# Split the dataset into training and testing sets
train_dataset, test_dataset = train_test_split(dataset, test_size=0.2, random_state=1)

print("Training examples: ", len(train_dataset))
print("Testing examples: ", len(test_dataset))

# Convert the datasets to TensorFlow Decision Forests compatible formats
train_dataset_tfdf = tfdf.keras.pd_dataframe_to_tf_dataset(train_dataset, label=label)
test_dataset_tfdf = tfdf.keras.pd_dataframe_to_tf_dataset(test_dataset, label=label)

def print_evaluation(evaluation, model_name):
    print(f"{model_name} Model Evaluation:")
    for key, value in evaluation.items():
        print(f"  {key}: {value}")

def get_accuracy(evaluation):
    # Attempt to find and return an appropriate accuracy metric
    for key in evaluation.keys():
        if 'accuracy' in key:
            return evaluation[key]
    return None

# Train and evaluate a Decision Tree (CART) model
cart_model = tfdf.keras.CartModel()
cart_model.fit(train_dataset_tfdf)
cart_train_evaluation = cart_model.evaluate(train_dataset_tfdf, return_dict=True)
cart_test_evaluation = cart_model.evaluate(test_dataset_tfdf, return_dict=True)
print_evaluation(cart_train_evaluation, "CART Train")
print_evaluation(cart_test_evaluation, "CART Test")
print("CART Test accuracy:", get_accuracy(cart_test_evaluation))

# Train and evaluate a Random Forest model
rf_model = tfdf.keras.RandomForestModel()
rf_model.fit(train_dataset_tfdf)
rf_train_evaluation = rf_model.evaluate(train_dataset_tfdf, return_dict=True)
rf_test_evaluation = rf_model.evaluate(test_dataset_tfdf, return_dict=True)
print_evaluation(rf_train_evaluation, "Random Forest Train")
print_evaluation(rf_test_evaluation, "Random Forest Test")
print("Random Forest Test accuracy:", get_accuracy(rf_test_evaluation))

# Train and evaluate a Gradient Boosted Trees model
gbt_model = tfdf.keras.GradientBoostedTreesModel()
gbt_model.fit(train_dataset_tfdf)
gbt_train_evaluation = gbt_model.evaluate(train_dataset_tfdf, return_dict=True)
gbt_test_evaluation = gbt_model.evaluate(test_dataset_tfdf, return_dict=True)
print_evaluation(gbt_train_evaluation, "Gradient Boosted Trees Train")
print_evaluation(gbt_test_evaluation, "Gradient Boosted Trees Test")
print("Gradient Boosted Trees Test accuracy:", get_accuracy(gbt_test_evaluation))

# Optimize the models by adjusting hyperparameters

# Hyperparameter tuning for CART model
print("Optimizing CART model:")
for max_depth in [3, 5, 10]:
    cart_model_tuned = tfdf.keras.CartModel(max_depth=max_depth)
    cart_model_tuned.fit(train_dataset_tfdf)
    cart_tuned_test_evaluation = cart_model_tuned.evaluate(test_dataset_tfdf, return_dict=True)
    print(f"CART Test accuracy with max_depth={max_depth}: {get_accuracy(cart_tuned_test_evaluation)}")

# Hyperparameter tuning for Random Forest model
print("Optimizing Random Forest model:")
for max_depth in [10, 20, None]:
    rf_model_tuned = tfdf.keras.RandomForestModel(max_depth=max_depth)
    rf_model_tuned.fit(train_dataset_tfdf)
    rf_tuned_test_evaluation = rf_model_tuned.evaluate(test_dataset_tfdf, return_dict=True)
    print(f"Random Forest Test accuracy with max_depth={max_depth}: {get_accuracy(rf_tuned_test_evaluation)}")

# Hyperparameter tuning for Gradient Boosted Trees model
print("Optimizing Gradient Boosted Trees model:")
for shrinkage in [0.01, 0.1, 0.2]:
    gbt_model_tuned = tfdf.keras.GradientBoostedTreesModel(shrinkage=shrinkage)
    gbt_model_tuned.fit(train_dataset_tfdf)
    gbt_tuned_test_evaluation = gbt_model_tuned.evaluate(test_dataset_tfdf, return_dict=True)
    print(f"Gradient Boosted Trees Test accuracy with shrinkage={shrinkage}: {get_accuracy(gbt_tuned_test_evaluation)}")
```

```python
# Import necessary libraries
import tensorflow_decision_forests as tfdf
import pandas as pd
import numpy as np

# Load the dataset
path = "https://storage.googleapis.com/download.tensorflow.org/data/palmer_penguins/penguins.csv"
dataset = pd.read_csv(path)
label = "species"

# Display the first 3 examples
print(dataset.head(3))

# Split the dataset into training and testing sets
np.random.seed(1)
is_test = np.random.rand(len(dataset)) < 0.2

train_dataset = dataset[~is_test]
test_dataset = dataset[is_test]

print("Training examples: ", len(train_dataset))
print("Testing examples: ", len(test_dataset))

# Convert pandas DataFrames to tfdf compatible datasets
train_dataset_tfdf = tfdf.keras.pd_dataframe_to_tf_dataset(train_dataset, label=label)
test_dataset_tfdf = tfdf.keras.pd_dataframe_to_tf_dataset(test_dataset, label=label)

# Train and evaluate the Random Forest model
print("\nTraining Random Forest model:")
rf_model = tfdf.keras.RandomForestModel()
rf_model.compile(metrics=["accuracy"])
rf_model.fit(train_dataset_tfdf)
rf_evaluation = rf_model.evaluate(test_dataset_tfdf, return_dict=True)
print("Random Forest test accuracy:", rf_evaluation["accuracy"])

# Train and evaluate the Gradient Boosted Trees model
print("\nTraining Gradient Boosted Trees model:")
gbt_model = tfdf.keras.GradientBoostedTreesModel()
gbt_model.compile(metrics=["accuracy"])
gbt_model.fit(train_dataset_tfdf)
gbt_evaluation = gbt_model.evaluate(test_dataset_tfdf, return_dict=True)
print("Gradient Boosted Trees test accuracy:", gbt_evaluation["accuracy"])

# Hyperparameter tuning for Random Forest
print("\nHyperparameter tuning for Random Forest:")

# Trying different values for max_depth
for max_depth in [3, 5, 10]:
    rf_model_tuned = tfdf.keras.RandomForestModel(max_depth=max_depth)
    rf_model_tuned.compile(metrics=["accuracy"])
    rf_model_tuned.fit(train_dataset_tfdf)
    rf_tuned_evaluation = rf_model_tuned.evaluate(test_dataset_tfdf, return_dict=True)
    print(f"Random Forest test accuracy with max_depth={max_depth}: {rf_tuned_evaluation['accuracy']}")

# Trying different values for num_trees
for num_trees in [50, 100, 200]:
    rf_model_tuned = tfdf.keras.RandomForestModel(num_trees=num_trees)
    rf_model_tuned.compile(metrics=["accuracy"])
    rf_model_tuned.fit(train_dataset_tfdf)
    rf_tuned_evaluation = rf_model_tuned.evaluate(test_dataset_tfdf, return_dict=True)
    print(f"Random Forest test accuracy with num_trees={num_trees}: {rf_tuned_evaluation['accuracy']}")

# Hyperparameter tuning for Gradient Boosted Trees
print("\nHyperparameter tuning for Gradient Boosted Trees:")

# Trying different values for max_depth
for max_depth in [3, 5, 10]:
    gbt_model_tuned = tfdf.keras.GradientBoostedTreesModel(max_depth=max_depth)
    gbt_model_tuned.compile(metrics=["accuracy"])
    gbt_model_tuned.fit(train_dataset_tfdf)
    gbt_tuned_evaluation = gbt_model_tuned.evaluate(test_dataset_tfdf, return_dict=True)
    print(f"Gradient Boosted Trees test accuracy with max_depth={max_depth}: {gbt_tuned_evaluation['accuracy']}")

# Trying different values for num_trees
for num_trees in [50, 100, 200]:
    gbt_model_tuned = tfdf.keras.GradientBoostedTreesModel(num_trees=num_trees)
    gbt_model_tuned.compile(metrics=["accuracy"])
    gbt_model_tuned.fit(train_dataset_tfdf)
    gbt_tuned_evaluation = gbt_model_tuned.evaluate(test_dataset_tfdf, return_dict=True)
    print(f"Gradient Boosted Trees test accuracy with num_trees={num_trees}: {gbt_tuned_evaluation['accuracy']}")
```
